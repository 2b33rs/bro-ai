```python
import pandas as pd
import csv
from dotenv import load_dotenv
import os

# .env-Datei laden
load_dotenv()
art_path = os.getenv("ART_PATH")
lv_path = os.getenv("LV_PATH")
rows_art = []
rows_lv = []
```

```python
# Robust einlesen, Zeilen mit Fehlern überspringen
with open(art_path, encoding='utf-8', errors='replace') as f:
    reader = csv.reader(f, delimiter=',', quotechar='"', doublequote=True)
    for i, row in enumerate(reader):
        if row:  # leere Zeilen überspringen
            rows_art.append(row)

# Optional: Spaltennamen setzen, falls CSV Header hat
columns = rows_art[0]  # erste Zeile als Header
data = rows_art[1:]    # restliche Zeilen

df_art = pd.DataFrame(data, columns=columns)

print(df_art.head(5))
print(f"DataFrame Größe: {df_art.shape}")
```

```python
# Robust einlesen, Zeilen mit Fehlern überspringen
with open(lv_path, encoding='utf-8', errors='replace') as f:
    reader = csv.reader(f, delimiter=',', quotechar='"', doublequote=True)
    for i, row in enumerate(reader):
        if row:  # leere Zeilen überspringen
            rows_lv.append(row)

# Optional: Spaltennamen setzen, falls CSV Header hat
columns = rows_lv[0]  # erste Zeile als Header
data = rows_lv[1:]    # restliche Zeilen

df_lv = pd.DataFrame(data, columns=columns)

print(df_lv.head(5))
print(f"DataFrame Größe: {df_lv.shape}")
```
